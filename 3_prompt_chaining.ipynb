```python
from langgraph.graph import StateGraph, END, START
from langchain_groq import ChatGroq
from typing import TypedDict
from dotenv import load_dotenv
import os
```

```python
load_dotenv()

GROQ_API_KEY = os.getenv('GROQ_API_KEY')

model = ChatGroq(
    model="llama-3.1-8b-instant",
    temperature=0
)
```

```python
class blog(TypedDict):
    title: str
    outline: str
    content: str
```

```python
def create_outline(state: blog) -> blog:
    title = state['title']

    result = model.invoke(f'write a outline on this topic \n {title}').content

    state['outline'] = result

    return state
```

```python
def create_blog(state: blog) -> blog:
    title = state['title']
    outline = state['outline']

    result = model.invoke(f'write a good blog on {title}, with this outline only \n {outline}').content

    state['content'] = result

    return state
```

```python
graph = StateGraph(blog)

graph.add_node('create_outline', create_outline)
graph.add_node('create_blog', create_blog)

graph.add_edge(START, 'create_outline')
graph.add_edge('create_outline', 'create_blog')
graph.add_edge('create_blog', END)

workflow = graph.compile()
```

```python
result = workflow.invoke({'title' : 'some good shayari with meaning of them'})
```
